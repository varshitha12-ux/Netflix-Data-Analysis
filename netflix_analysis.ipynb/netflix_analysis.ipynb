# Netflix Data Analysis

## Objective

This project analyzes Netflix Movies and TV Shows data using Python and Matplotlib.

### Goals:
- Understand content distribution
- Find top genres
- Identify top content-producing countries
- Analyze content growth over years

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("dataset/netflix_titles.csv")
```

```python
df_clean = df.copy()
```

```python
print("Dataset Shape After Cleaning:")
df_clean.shape
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean = df.copy()
```

```python
df_clean = df_clean.fillna("Unknown")
```

```python
df_clean.isnull().sum()
```

```python
df_clean.shape
```

```python
print("type_counts:")
df_clean["type"].value_counts()
```

```python
df_clean["type"].value_counts().plot(
    kind="bar",
    title="Movies vs TV Shows on Netflix",
    xlabel="content type",
    ylabel="count",
    )
plt.savefig="movies_vs_tvshows.png"
plt.show()
```

```python
country_count=df_clean["country"].value_counts().head(10)
print(country_count)
```

```python
df_clean["country"].value_counts().head(10).plot(
    kind="bar",
    xlabel="Country",
    ylabel="no.of titles",
    title="Top Netflix Countries")
plt.savefig="top_countries.png"
plt.show()
```

```python
genre_count=df_clean["listed_in"].value_counts().head(10)
print(genre_count)
```

```python
df_clean["listed_in"].value_counts().head(10).plot(
    kind="bar",
    xlabel="Genre",
    ylabel="Number of Titles",
    title="Top Netflix Genres")
plt.savefig="top_genres.png"
plt.show()
```

```python
year_count = df_clean["release_year"].value_counts().sort_index()
print(year_count.tail())
```

```python
df_clean["release_year"].value_counts().sort_index().plot(
    kind="line",
    xlabel="Release Year",
    ylabel="Number of Titles",
    title="Netflix Content Growth Over Years")
plt.savefig="release_trend.png"
plt.show()
```

```python
print("Key Findings:")
print("1. Movies are more common than TV Shows.")
print("2. USA produces the highest Netflix content.")
print("3. Drama-related genres dominate.")
print("4. Netflix content growth accelerated after 2015.")
```

Conclusion

- Movies are more common than TV Shows.
- USA produces the most content.
- Drama genres dominate Netflix.
- Content growth increased significantly after 2015.
